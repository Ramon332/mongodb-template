```python
dados_mongo = list(colecao.find({
    'valor' : {'$gte' : 90, '$lte': 200}
}))
display(dados_mongo)
```

```python
dados_cidade = list(colecao.find({
    'cidade' : {'$in' : ['Ourinhos', 'Assis', 'Pompeia', 'Marília'],}
}))
display(dados_cidade)
```

```python
df = pd.DataFrame(dados_cidade)
df
```

```python
df.isnull().sum()
```

```python
faturamento = df['valor'].sum()
print("Faturamento Total: ", faturamento)
```

```python
faturamento_cidade = df.groupby('cidade')['valor'].sum().sort_values(ascending = False)
faturamento_cidade
```

```python
faturamento_servico = df.groupby('servico')['valor'].sum().sort_values(ascending = False)
faturamento_servico
```

```python
ticket_medio = df['valor'].mean()
print(f"Média geral do ticket: {ticket_medio:.2f}")
```

```python
ticket_medio_cidade = df.groupby('cidade')['valor'].mean().sort_values(ascending = False)
ticket_medio_cidade
```

```python
servico_lucrativo = df.groupby('servico')['valor'].sum().idxmax()
servico_lucrativo
```

```python
cidade_maior_lucro = df['cidade'].value_counts().idxmax()
cidade_maior_lucro
```

```python
top_3_atendimentos = df['servico'].value_counts().head(3)
print("Top 3 atendimentos com mais atendimentos: ")
print(top_3_atendimentos)
```

```python
relatorio_atendimento = {
    'data_geracao' : datetime.now(),
    'faturamento_total' : float(faturamento),
    'faturamento_cidade' : faturamento_cidade.to_dict(),
    'faturamento_servico' : faturamento_servico.to_dict(),
    'ticket_medio' : ticket_medio,
    'ticket_medio_cidade' : ticket_medio_cidade.to_dict(),
    'servico_lucrativo' : servico_lucrativo,
    'cidade_maior_lucro' : cidade_maior_lucro,
    'Top_3_atendimentos' : top_3_atendimentos.to_dict()
}
display(relatorio_atendimento)
```

```python
colecao_relatorios = db['relatorios']
colecao_relatorios.insert_one(relatorio_atendimento)
```
